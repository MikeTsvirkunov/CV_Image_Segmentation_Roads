```python
import pandas as pd
import cv2
import seaborn_image as isns
from IPython.display import Markdown as md
from os import listdir
from os.path import isfile, join
import numpy as np
```

```python
standart_image_path = './data/images/0.png'
segmented_image_path = './data/masks/0.png'
standart_image = cv2.cvtColor(cv2.imread(standart_image_path), cv2.COLOR_BGR2RGB)
segmented_image = cv2.cvtColor(cv2.imread(segmented_image_path), cv2.COLOR_BGR2RGB)
isns.imgplot(standart_image)
isns.imgplot(segmented_image)
```

```python
hsv_v = cv2.cvtColor(standart_image, cv2.COLOR_RGB2HSV)
rgb_v = cv2.cvtColor((hsv_v * np.array([1, 1.5, 1])).astype(np.uint8), cv2.COLOR_HSV2RGB)
isns.imgplot(standart_image, )
isns.imgplot(rgb_v, )
```

# Create dataset for load img

```python
def getFileListInDir(path):
    return sorted([path + f for f in listdir(path) if isfile(join(path, f))])
```

```python
images_path = './data/images/'
masks_path = './data/masks/'
dataset = pd.DataFrame({'images': getFileListInDir(images_path), 'masks': getFileListInDir(masks_path)})
dataset.to_csv('data/images_masks_pathes_pairs.csv', index=False)
dataset.head()
```
